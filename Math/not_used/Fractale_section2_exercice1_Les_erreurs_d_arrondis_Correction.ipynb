#### 1- a- Calculer à la main l’écriture binaire des nombres suivants  : 0.5, 0.25 et 0.1

Comment faire ? Pour un entier c’est chose facile : $3 = 2¹ + 2⁰$  donc son écriture binaire est 11. Pour un nombre à virgule, on procède de la même manière : $0.625 = 2⁻¹ + 2⁻³$ donc son écriture binaire est 0.101.


0.5 = 2⁻¹ on en déduit donc que son écriture binaire est 0.1. De la même manière, 0.25 = 2⁻² donc son écriture binaire est 0.01.
Pour 0.1 (ici écrit en base 10), c’est plus compliqué… 0.1 = 2⁻⁴+ 2⁻⁵ + 2⁻⁸ + 2⁻⁹ + 2⁻¹² + 2⁻¹³ … donc l’écriture binaire de 0.1 (base 10) est 0.0001100110011… en binaire.



 ####  b- Quelle est la particularité de 0.1 par rapport à 0.5 ou 0.25 ?

L’écriture binaire de 0.1 est en fait infinie, un ordinateur ne peut donc pas avoir une représentation exacte de ce nombre pourtant si évident en base 10.

#### D’où viennent les erreurs ? Pourquoi celles-ci deviennent-elles si importantes au bout de quelques itérations seulement ?



Ces erreurs viennent encore une fois de l’impossibilité de représenter correctement 0.1 en binaire, et donc par conséquent de représenter correctement A et B. Au début, les erreurs des calculs en 64 bits sont minimes : x1 est estimé à 1.0000000000004547 l’erreur est donc de l’ordre de 10⁻¹³ à l’itération suivante, l’erreur est de l’ordre de 10⁻⁹, à l’itération 3 de 10⁻⁵… L’ordre de grandeur de l’erreur est multiplié par 10⁴ à chaque itération. Cela est tout simplement dû au fait que A est de l’ordre de 10⁴ :
$x_{n+1} = A * x_n -B= A * 1 - B + A * Erreur_n $
or la valeur exacte de $x_{n+1}$ est $ A * 1 - B = 1 $puisque la valeur exacte de xn est 1 donc l’erreur sur l’estimation de $x_n+1$ est $Erreur_{n+1}=A * Erreur_n$.
l’erreur est ainsi multiplié par A (donc par un ordre de grandeur de 10⁴) à chaque itération.
