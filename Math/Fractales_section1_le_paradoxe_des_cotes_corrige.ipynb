```python
#Paradoxe des cotes corrige

#Dans ce corrig, nous compltons le programme avec les valeurs QUE NOUS AVONS MESUREES. Il est possible que vos valeurs soient diffrentes !
nombre_de_points=[2,4,8,16,32,64,128,256]

distance=[90,94,100,112,124,144,160,179]


import numpy as np
import math as math
import matplotlib.pyplot as plt
import time


#Le paradoxe des cotes


nombre_de_points=np.zeros(8) #Nombre de points utiliss  chaque tape

distance=np.zeros(8) #distances mesures entre Saint Brieuc et le Mont-Saint-Michel  chaque tape


##################################### QUESTION 1 #############################################


nombre_de_points[0]=2 #Completez avec le nombre de points utiliss  l'tape 1

distance[0]=90 #Completez la distance totale que vous mesurez  l'tape 1




##################################### QUESTION 2 #############################################


nombre_de_points[1]=4 #Completez avec le nombre de points utiliss  l'tape 2

distance[1]=94 #Completez la distance totale que vous mesurez  l'tape 2




##################################### QUESTION 3 #############################################


nombre_de_points[2]=8 #Completez avec le nombre de points utiliss  l'tape 3

distance[2]=100 #Completez la distance totale que vous mesurez  l'tape 3




##################################### QUESTION 4 #############################################


nombre_de_points[3]=16 #Completez avec le nombre de points utiliss  l'tape 4

distance[3]=112 #Completez la distance totale que vous mesurez  l'tape 4



##################################### QUESTION 5 #############################################


nombre_de_points[4]=32 #Completez avec le nombre de points utiliss  l'tape 5

distance[4]=124 #Completez la distance totale que vous mesurez  l'tape 5


nombre_de_points[5]=64 #Completez avec le nombre de points utiliss  l'tape 6

distance[5]=144 #Completez la distance totale que vous mesurez  l'tape 6


nombre_de_points[6]=128 #Completez avec le nombre de points utiliss  l'tape 7

distance[6]=160 #Completez la distance totale que vous mesurez  l'tape 7


nombre_de_points[7]=256 #Completez avec le nombre de points utiliss  l'tape 8

distance[7]=179 #Completez la distance totale que vous mesurez  l'tape 8



##############################################################################################

#Affichage

plt.rcParams['figure.figsize'] = [14, 14] #Redimensionne les figures

a, b = np.polyfit(np.log(nombre_de_points), np.log(distance), 1) #Calcul de la droite la plus proche de la mesure


plt.plot(nombre_de_points,distance, label='Distance totale mesure')

# Affiche la grille
plt.grid(True, which="both")

# Titre
plt.title('Distance totale mesuree entre St Brieuc et le Mont Saint-Michel en fonction du nombre de points de mesure')

# Axe x
plt.xlabel('Nombre de points')
 
# Axe y
plt.ylabel('Distance totale mesuree')

# Affichage
plt.show()





fig, ax = plt.subplots()

ax.plot(np.log(nombre_de_points), np.log(distance), label='Mesure') #Affiche les points mesurs
ax.plot(np.log(nombre_de_points), np.multiply(np.log(nombre_de_points),a)+b, label='Droite interpolee (a='+str(round(a,3))+')') #Affiche la droite interpole



# Affiche la grille
plt.grid(True, which="both")

# Titre
plt.title('log(distance) en fonction de log(nombre_de_points)')

# Axe x
plt.xlabel('log(Nombre de points)')

# Axe y
plt.ylabel('log(Distance totale mesuree)')

legend = ax.legend(loc='upper left', shadow=True, fontsize='x-large')

# Affichage
plt.show()


```
