Le premier et le troisime point sont faciles  placer. Ils se situent sur l'axe entre les deux extrmit divisant le segment en rapport sous-segments de mme longueur.

Pour le deuxime point, il se trouve  mi-chemin entre les deux segments mais dcal de HAUTEUR orthogonalement au segment. HAUTEUR se calcule de la faon suivante :

HAUTEUR = sin(angle)/rapport

Ainsi, les coordonnes des points B, C, C', et D sont les suivantes :

$(B_x, B_y) = (A_x + AE_x\_reduit, A_y + AE_y\_reduit) $

$(C'_x, C'_y) = (Ax + (AE / MILIEU)_x, A_t + (AE / MILIEU)_y)$

Remplissez maintenant la fonction placer_sommet en vous aidant de ces formules. Vous noterez que celles-ci ne sont plus valide si les coordonnes de A et E sont changes (cela retourne le triangle). Il faudra donc s'assurer que indice1 < indice2.

```python
def placer_sommets(indice1, indice2, angle, rapport) :
    """
    Prrequis :
    indice1 et indice2 sont des entiers positifs strictements infrieurs 
    len(abscisses_tab)=len(ordonnees_tab). indice1 < indice2 et 
    |indice2 - indice1| - 1 est une puissance de trois. Cette 
    dernire condition assure qu'il reste bien  placer un nombre de 
    sommets cohrent avec la courbe de Von Koch.
    angle est un flottant et est exprim en radians.
    rapport est un entier indiquant le rapport d'omothtie entre deux itrations
    successives (pour la courbe de Von Koch il vaut 3).

    
    Remplit les coordonnes des sommets d'indice compris entre indice1 et indice2
    dans les tableaux abscisses_tab et ordonnees_tab en respectant l'angle angle.
    
    Ne renvoie rien.
    """
    # On vrifie que indice1 et indice2 sont bien des indices 
    # d'lments d'abscisses_tab et d'ordonnees_tab
    assert(indice1 >= 0 and indice1 < len(abscisses_tab))
    assert(indice2 >= 0 and indice2 < len(abscisses_tab))
    
    # On vrifie que indice1 < indice2
    assert(indice1 < indice2)
    
    # On vrifie que le nombre de sommets entre indice1 et indice2
    # est bien trois.
    assert((indice2 - indice1 - 1) == 3)
    
    # On utilisera les notations de la figure ci-dessus.   
    MILIEU = 2
    HAUTEUR = sin(angle)/rapport
    
    # Rcupration des coordonnes connues
    Ax = abscisses_tab[indice1]
    Ay = ordonnees_tab[indice1]
    Ex = abscisses_tab[indice2]
    Ey = ordonnees_tab[indice2]
    
    # "hauteur" du point C sur l'axe (AE)
    Hx=(Ex+Ax)/MILIEU
    Hy=(Ey+Ay)/MILIEU
    
    # Calcul du vecteur AB
    AEx_reduit = (Ex-Ax) / rapport
    AEy_reduit = (Ey-Ay) / rapport

    # Calcul du vecteur C'C
    HCx =(Ey-Ay) * HAUTEUR
    HCy =(Ex-Ax) * HAUTEUR

    # Coordonnes de B
    Bx = Ax + AEx_reduit
    By = Ay + AEy_reduit

    # Coordonnes de C
    Cx = Hx - HCx
    Cy = Hy + HCy
    
    # Coordonnes de D
    Dx = Ex - AEx_reduit
    Dy = Ey - AEy_reduit

    # Pour l'instant d vaut 1 
    d = int((indice2 - indice1) / 4)
    
    # Assignation des coordonnes calcules dans le tableau

    abscisses_tab[indice1 + 1 * d] = Bx
    ordonnees_tab[indice1 + 1 * d] = By

    abscisses_tab[indice1 + 2 * d] = Cx
    ordonnees_tab[indice1 + 2 * d] = Cy

    abscisses_tab[indice1 + 3 * d] = Dx
    ordonnees_tab[indice1 + 3 * d] = Dy
```
